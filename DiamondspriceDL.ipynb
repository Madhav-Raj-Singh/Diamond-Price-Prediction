```python
import seaborn as sns
df=sns.load_dataset('diamonds')
df.head()
```

```python
import pandas as pd
import numpy as np
```

```python
X = df.iloc[:, [0,1,2,3,4,5,7,8,9]]
y = df.iloc[:, 6]
```

```python
cut=pd.get_dummies(X['cut'],drop_first=True).astype(int)
color=pd.get_dummies(X['color'],drop_first=True).astype(int)
clarity=pd.get_dummies(X['clarity'],drop_first=True).astype(int)
```

```python
X=X.drop(['cut','color','clarity'],axis=1)
```

```python
X=pd.concat([X,cut,color,clarity],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import ReLU,LeakyReLU
from tensorflow.keras.layers import Dropout
```

```python
classifier=Sequential()

classifier.add(Dense(64,activation='relu'))
classifier.add(Dropout(0.3))

classifier.add(Dense(30,activation='relu'))
classifier.add(Dropout(0.2))

classifier.add(Dense(1,activation='linear'))

```

```python
classifier.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
early_stoping= tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=True,
    start_from_epoch=0,
)
```

```python
classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=1000,callbacks=[early_stoping])
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
